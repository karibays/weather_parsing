```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather_df = pd.read_excel('dataset/april_weather.xls')
```

```python
d = {}
for i, j in zip(weather_df.columns, weather_df.loc[5]):
    d[f'{i}'] = j
```

```python
weather_df = weather_df.rename(columns=weather_df.loc[5])
weather_df = weather_df.rename(columns={'Местное время в Астане': 'date'})
```

```python
weather_df.info()
```

```python
weather_df = weather_df.drop(columns=['ff10', 'ff3', 'W1', 'W2',
                        'Tn', 'Tx', 'Cl', 'Nh',
                        'H', 'Cm', 'Ch', 'VV',
                        'RRR', 'tR', 'E', 'Tg',
                        "E'", 'sss', 'WW', 'DD'])
```

```python
weather_df = weather_df.drop(index=[0,1,2,3,4,5])
weather_df = weather_df.reset_index(drop=True)
weather_df['P'] = weather_df['P'].fillna(weather_df['P'].astype('float32').mean())

weather_df.head(5)
```

```python
weather_df.info()
```

```python
# all_DD_features = weather_df['Unnamed: 6'].astype("category").cat.categories
# all_DD_features = {item: i + 1 for i, item in enumerate(weather_df['DD'].unique())}

# weather_df['DD'] = weather_df['DD'].map(all_DD_features)
```

```python
# weather_df['N'].unique()
weather_df['N'] = weather_df['N'].replace('Облаков нет.', 0)
weather_df['N'] = weather_df['N'].replace(['20–30%.', '70 – 80%.', '90  или более, но не 100%'], 
                                                              [25, 75, 95])

weather_df['N'] = weather_df['N'].replace(r'\D+', '', regex=True).astype('float16')
```

```python
weather_df.head(5)
```

```python
weather_df['date'] = pd.to_datetime(weather_df['date'])
```

```python
weather_df['year'] = weather_df['date'].dt.year
weather_df['month'] = weather_df['date'].dt.month
weather_df['day'] = weather_df['date'].dt.day
weather_df['hour'] = weather_df['date'].dt.hour
weather_df['dayofweek'] = weather_df['date'].dt.dayofweek
weather_df = weather_df.drop(columns='date')

weather_df.head(5)
```

```python
weather_df = weather_df.astype('float16')
weather_df
```

```python
weather_df.head(5)
```

```python
weather_df.describe()
```

```python
weather_df['U'] = weather_df['U'].astype('uint8')
weather_df['Ff'] = weather_df['Ff'].astype('uint8')
weather_df['N'] = weather_df['N'].astype('uint8')

weather_df['year'] = weather_df['year'].astype('uint16')
weather_df['month'] = weather_df['month'].astype('uint8')
weather_df['day'] = weather_df['day'].astype('uint8')
weather_df['hour'] = weather_df['hour'].astype('uint8')
weather_df['dayofweek'] = weather_df['dayofweek'].astype('uint8')
```
